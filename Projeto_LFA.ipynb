```python
def create_dfa_step_image(current_state, input_symbol, next_state, step,digraph):
    dot = Digraph()

    all_states = ['q0', 'q1', 'q2', 'q3', 'q4', 'q5', 'q6', 'q7', 'q8', 'q9',
     'q10', 'q11', 'q12', 'q13', 'q14', 'q15', 'q16', 'q17', 'q18', 'q19',
     'q20', 'q21', 'q22', 'q23', 'q24', 'q25', 'q26', 'q27', 'q28', 'q29',
     'q30', 'q31', 'q32', 'q33', 'q34', 'q35', 'q36', 'q37', 'q38', 'q39',
     'q40', 'q41', 'q42', 'q43', 'q44', 'q45', 'q46', 'q47', 'q48', 'q49',
     'q50',        'q52',        'q54', 'q55', 'q56', 'q57', 'q58', 'q59',
     'q60', 'q61', 'q62', 'q63', 'q64', 'q65', 'q66', 'q67']

    # Verifica se Estados são Finais e Desenha o Aro Dupli caso Positivo
    if checkFinal(current_state,F):
      #print("final current")
      dot.node(current_state, peripheries='2')
      digraph.node(current_state, peripheries='2')
    else:
      dot.node(current_state)
      digraph.node(current_state)


    if checkFinal(next_state,F):
      #print("final next\n")
      dot.node(next_state, peripheries='2')
      digraph.node(next_state, peripheries='2')
    else:
      dot.node(next_state)
      digraph.node(next_state)

    # Ajusta os Parametros dos Estados
    dot.attr('node', style='filled', color='lightgrey')
    dot.node(current_state)  # Estado Atual
    dot.node(next_state)  # Proximo Estado
    dot.attr('node', style='', color='')  # Reseta para o padrão

    digraph.attr('node', style='filled', color='lightgrey')
    digraph.node(current_state)
    digraph.node(next_state)
    digraph.attr('node', style='', color='')

    # Adiciona Transição
    #print(f"current={current_state} next={next_state}")
    dot.edge(current_state, next_state, label=' '+ input_symbol + f' [{step}]', color='red', penwidth='2.0')
    digraph.edge(current_state, next_state, label=' '+ input_symbol + f' [{step}]', color='red', penwidth='2.0')

    # Gera e Salva Imagem
    filename1 = f'./content/dfa_image'
    filename2 = f'./content/dfa_step_{step}'
    digraph.render(filename1, format='png', view=False)
    dot.render(filename2, format='png', view=False)
    #return filename2 + '.png'  # Return the full path to the generated image
    return digraph
# Generate an image for a step
#image_path = create_dfa_step_image('q0', 'a', 'q2', True, 1)

# Display the image
#display(Image(filename=image_path))
```

```python
def afdConstruct(M, w):
    digraph = Digraph()
    # Recebe os elementos da Tupla
    δ, q, F = M
    step = 0
    # Itera sobre cada símbolo da palavra w
    for s in w:
        # Check if the current state and symbol combination exists in the transition function
        #Checa se a combinaçaõ (estado atual, simbolo) existe na função de Transição δ
        if (q, s) in δ:
            q2 = δ[(q, s)]  # Recebe o Proximo Estado
            #print(f"s={s} q={q} q2={q2} check={checkFinal(q2,F)}")  # Optional: print the current symbol and state
            create_dfa_step_image(q, s, q2, step,digraph) # Cria a imagem do Caminho Percorrido
            step += 1 # Atualiza a Iteração
            q=q2 #Atualiza o caracter atual
        else:
            print(f"Invalid transition: No transition defined for state {q} with input {s}")
            return False  # Função de Saida para valores invalidos

    # Check if the final state is in the set of accepting states
    #print(q2 in F)
    return dfa.input_check(w), q2, step # Retorna uma Tabela, Estado Final e Número de Iterações
    #return q2 in F
```

```python
from PIL import Image
def conc2img(path_img1, path_img2): # Usado para Concatenar as Duas Ultimas Imagens

  image1 = Image.open(path_img1)
  image2 = Image.open(path_img2)
  total_width = image1.width + image2.width
  max_height = max(image1.height, image2.height)

  new_im = Image.new('RGB', (total_width, max_height))

  new_im.paste(image1, (0, 0))
  new_im.paste(image2, (image1.width, 0))

  new_im.save('final_h.png')
```
